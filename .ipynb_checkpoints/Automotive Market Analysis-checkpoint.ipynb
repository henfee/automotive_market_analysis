```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import quandl
import datetime
%matplotlib inline
```

```python
start = "2012-1-1"
end = "2017-1-1"
```

```python
dataSource = "quandl"
```

```python
tsla = quandl.get(dataset="WIKI/TSLA", start_date=start, end_date=end)
ford = quandl.get(dataset="WIKI/F", start_date=start, end_date=end)
gm = quandl.get(dataset="WIKI/GM", start_date=start, end_date=end)
```

#### Visualization

```python
tsla["Open"].plot(label="Tesla", figsize=(16, 8), title="Open Price")
ford["Open"].plot(label="Ford")
gm["Open"].plot(label="GM")
plt.legend()
```

```python
tsla["Volume"].plot(label="Tesla", figsize=(16, 8), title="Open Price")
ford["Volume"].plot(label="Ford")
gm["Volume"].plot(label="GM")
plt.legend()
```

```python
ford["Volume"].argmax()
```

##### The volume of shares traded spikes for Ford's stock on December 12, 2013 becasue their 2014 sales were projected to fall.

##### Normalizing for market cap..

##### "Total Traded" is the total amount of cash exchanged

```python
stocks = [tsla, ford, gm]
for stock in stocks:
    stock["Total Traded"] = stock["Open"] * stock["Volume"]
```

```python
tsla["Total Traded"].plot(label="Tesla", figsize=(16, 8), title="Open Price")
ford["Total Traded"].plot(label="Ford")
gm["Total Traded"].plot(label="GM")
plt.legend()
plt.ylabel("Total Traded")
```

```python
tsla["Total Traded"].argmax()
```

##### On Feburary 25, 2014, Tesla reported a higher than expected profit.

### Candlestick Charts

```python
from matplotlib.finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY
```

```python
def candlestick(stocks):
    mondays = WeekdayLocator(MONDAY)
    alldays = DayLocator()
    weekFormatter = DateFormatter("%b %d")
    dayFormatter = DateFormatter("d")
    
    
    
    f, axarr = plt.subplots(nrows=len(stocks), figsize=(12, 18), sharex=True)
    st = f.suptitle("Candle Stick Plots of TSLA, FORD, and GM", fontsize="x-large")
    st.set_y(0.90)
    f.subplots_adjust(bottom=0.2)
    
    for index, stock in enumerate(stocks):
        _reset = stock.loc["2012-01":"2012-01"].reset_index()
        _reset["date_ax"] = _reset["Date"].apply(lambda date: date2num(date))
        _values = [tuple(vals) for vals in _reset[["date_ax", "Open", "High", "Low", "Close"]].values]
        axarr[index].xaxis.set_major_locator(mondays)
        axarr[index].xaxis.set_minor_locator(alldays)
        axarr[index].xaxis.set_major_formatter(weekFormatter)
        candlestick_ohlc(axarr[index], _values, width=0.6, colorup='g',colordown='r')
```

```python
candlestick([tsla, ford, gm])
```

## Financial Analysis

```python
for stock in stocks:
    stock["returns"] = stock["Close"] / (stock["Close"]).shift(1) - 1
```

```python
tsla.head()
```

```python
tfg_returns = pd.concat([tsla["returns"], ford["returns"], gm["returns"]], axis=1)
```

```python
tfg_returns.columns = ["Tesla Returns", "Ford Returns", "GM Returns"]
```

```python
tfg_returns.dropna(inplace=True) # No returns on the first day
```

```python
tfg_returns.head()
```

```python
tfg_returns.plot(kind="hist", bins=150, alpha=0.5, figsize=(16, 8))
```

```python
tfg_returns.plot(kind="kde", alpha=0.5, figsize=(16, 8))
```

```python
tfg_returns.plot(kind="box", figsize=(16, 8), colormap="jet")
```

#### Correlations between returns

```python
scatter_matrix(tfg_returns, figsize=(16, 16), hist_kwds={"bins": 50});
```

```python
tfg_returns.plot(kind="scatter", x="GM Returns", y="Ford Returns", figsize=(12, 6), alpha=0.5)
```

### Cumulative Daily Return

```python
for stock in stocks:
    stock["Cumulative Return"] = (1 + stock["returns"]).cumprod()
```

```python
tsla.head()
```

```python
tfg_cum_return = pd.concat([tsla["Cumulative Return"], ford["Cumulative Return"], gm["Cumulative Return"]], axis=1)
```

```python
tfg_cum_return.dropna(inplace=True)
tfg_cum_return.columns = ["Tesla Cumulative Return", "Ford Cumulative Return", "GM Cumulative Return"]
```

```python
tfg_cum_return.plot(figsize=(16, 8))
```

##### F and GM show similar cumulative returns, whereas TSLA's cumulative returns are larger.
